```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generar datos para la distribución normal
media = 70
desv_std = 4
datos = np.random.normal(media, desv_std, 1000)

# Crear el rango de valores para la curva PDF
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 100)
pdf = norm.pdf(x, media, desv_std)
```

```python
# Graficar histograma
plt.figure(figsize=(6, 4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color='b')
plt.plot(x, pdf, 'k', linewidth=2)

plt.title('Distribución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```
