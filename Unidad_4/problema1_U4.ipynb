```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Parámetros del problema
media = 70
desv_std = 4
total_estudiantes = 500
limite_inferior = 68
limite_superior = 72

# 2. Cálculo de probabilidades
# La probabilidad entre dos valores es: P(X < 72) - P(X < 68)
prob_68_72 = norm.cdf(limite_superior, media, desv_std) - norm.cdf(limite_inferior, media, desv_std)

# 3. Cálculo del número de estudiantes
cantidad_estudiantes = total_estudiantes * prob_68_72

# Mostrar resultados
print(f"1. La probabilidad de pesar entre 68 y 72 kg es: {prob_68_72:.4f} (o {prob_68_72*100:.2f}%)")
print(f"2. El número de estudiantes esperado es: {cantidad_estudiantes:.2f} (aproximadamente {round(cantidad_estudiantes)} estudiantes)")

# --- Visualización ---
plt.figure(figsize=(8, 5))

# Generar puntos para la curva
x = np.linspace(media - 4*desv_std, media + 4*desv_std, 1000)
y = norm.pdf(x, media, desv_std)

# Dibujar la curva normal
plt.plot(x, y, 'b', linewidth=2, label='Distribución Normal')

# Sombrear el área de interés (entre 68 y 72)
x_sombra = np.linspace(limite_inferior, limite_superior, 100)
y_sombra = norm.pdf(x_sombra, media, desv_std)
plt.fill_between(x_sombra, y_sombra, color='skyblue', alpha=0.5, label='Área buscada (68-72 kg)')

# Detalles de la gráfica
plt.title('Distribución de Pesos de Estudiantes ITD')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.axvline(media, color='red', linestyle='--', label=f'Media ({media} kg)')
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.show()
```
