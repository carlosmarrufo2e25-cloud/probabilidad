###problema 1

```python
from scipy.stats import poisson
# la media de accidentes por semana
media_semanal=3

# a) Probabilidad de que no ocurra ningún accidente en una semana
# aqui x=0
prob_ninguno = poisson.pmf(0, media_semanal)

# b) Probabilidad de que ocurran como máximo 2 accidentes en una semana
prob_maximo_2 = poisson.cdf(2, media_semanal)

# c) Probabilidad de que ocurran 10 accidentes en 3 semanas
# se ajusta la media para el periodo de 3 semanas: 3*3=9
media_tres_semanas = media_semanal * 3
# aqui x=10
prob_diez_en_tres = poisson.pmf(10, media_tres_semanas)

# Los resultados
print(f"a) Probabilidad de 0 accidentes en una semana: {prob_ninguno:.4f}")
print(f"b) Probabilidad de máximo 2 accidentes en una semana: {prob_maximo_2:.4f}")
print(f"c) Probabilidad de 10 accidentes en 3 semanas: {prob_diez_en_tres:.4f}")
```

###problema 2

```python
from scipy.stats import poisson
# el promedio de imperfecciones por milímetro
media_imperfecciones = 2

# a) Probabilidad de que no haya imperfecciones en 1 milímetro de alambre
# x=0
prob_a = poisson.pmf(0, media_imperfecciones)

# b) Probabilidad de que haya como máximo 3 imperfecciones en 1 milímetro
prob_b = poisson.cdf(3, media_imperfecciones)

# c) Probabilidad de que haya 5 imperfecciones en 2 milímetros de alambre
# se ajusta la media para 2 mm: 2 imperfecciones/mm * 2 mm = 4
media_2mm = media_imperfecciones * 2
# x=5
prob_c = poisson.pmf(5, media_2mm)

# Resultados
print(f"a) Probabilidad de 0 imperfecciones en 1 mm: {prob_a:.4f}")
print(f"b) Probabilidad de máximo 3 imperfecciones en 1 mm: {prob_b:.4f}")
print(f"c) Probabilidad de 5 imperfecciones en 2 mm: {prob_c:.4f}")
```
