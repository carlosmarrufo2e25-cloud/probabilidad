```python
from statistics import *
```

```python
# lista de calificaciones del grupo a de probabilidad y estadistica
grupoa = [70,70,65,90,100,95,45,80,73,20,100,96]
print(grupoa)
```

```python
# sacar la media de las calificaciones con la formula
promedio = sum(grupoa) / len(grupoa)
print(promedio)
```

```python
#determinar el valor de la media con la funcion median
mediana = median(grupoa)
print(mediana)
```

```python
# determinar la moda
moda = mode(grupoa)
print(moda)
```

```python
#determinar la multimoda
moda2 = multimode(grupoa)
print(moda2)
```

```python
#para crear numeros random
import random
```

```python
#fabrica 1
fabrica1 =[]
for i in range(100):
  numero_aleatoria= round(random.uniform(50,100), 2)
  fabrica1.append(numero_aleatoria)
print(fabrica1)
```

```python
#fabrica 2
fabrica2 =[]
for i in range(100):
  numero_aleatorio= round(random.uniform(70,100), 2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
# libreria de python para graficar valores
import matplotlib.pyplot as plt
```

```python
#imprimir grafica 1
plt.hist(fabrica1)
plt.title("produccion diaria de la fabrica 1")
```

```python
#imprimir grafica 2
plt.hist(fabrica2)
plt.title("produccion diaria de la fabrica 2")
```

```python
print("valores de la fabrica 1 ")
print("los valores de la fabrica 1 es =>", mean(fabrica1) )
print("la moda de la fabrica 1 es =>", mode(fabrica1) )
print("la media de la fabrica 1 es =>", median(fabrica1) )
```

```python
print("valores de la fabrica 2 ")
print("los valores de la fabrica 2 es =>", mean(fabrica2) )
print("la moda de la fabrica 2 es =>", mode(fabrica2) )
print("la media de la fabrica 2 es =>", median(fabrica2) )
```
