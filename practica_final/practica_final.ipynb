```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

from google.colab import drive

# =========================
# CONECTAR GOOGLE DRIVE
# =========================

drive.mount('/content/drive', force_remount=True)

# =========================
# CARGAR ARCHIVO CSV
# =========================

ruta_datos = '/content/drive/MyDrive/Colab Notebooks/covid2020-2021-codificado.csv'

df = pd.read_csv(ruta_datos, low_memory=False)

# =========================
# LIMPIAR COLUMNAS
# =========================

df.columns = df.columns.str.upper().str.strip()

print("Primeros registros del dataset:")
print(df.head())

# =========================
# DETECTAR COLUMNAS
# =========================

col_clasificacion = 'CLASIFICACION_FINAL' if 'CLASIFICACION_FINAL' in df.columns else None

col_paciente = 'TIPO_PACIENTE' if 'TIPO_PACIENTE' in df.columns else None

col_embarazo = 'EMBARAZO' if 'EMBARAZO' in df.columns else None

col_indigena = 'INDIGENA' if 'INDIGENA' in df.columns else None

col_entidad = 'ENTIDAD_RES' if 'ENTIDAD_RES' in df.columns else None

col_fecha = 'FECHA_INGRESO' if 'FECHA_INGRESO' in df.columns else None

col_sexo = 'SEXO' if 'SEXO' in df.columns else None

col_edad = 'EDAD' if 'EDAD' in df.columns else None

# =========================
# ESTADÍSTICA DESCRIPTIVA
# =========================

print("\n" + "="*60)
print(" ESTADÍSTICA DESCRIPTIVA ")
print("="*60)

# -------------------------
# CLASIFICACIÓN DE CASOS
# -------------------------

if col_clasificacion:
    print("\n[Casos por Clasificación Final]")
    print(df[col_clasificacion].value_counts())

# -------------------------
# TIPO DE PACIENTE
# -------------------------

if col_paciente:
    print("\n[Pacientes Ambulatorios y Hospitalizados]")
    print(df[col_paciente].value_counts())

# -------------------------
# EMBARAZO
# -------------------------

if col_embarazo:
    print("\n[Pacientes Embarazadas]")
    print(df[col_embarazo].value_counts())

# -------------------------
# INDÍGENAS
# -------------------------

if col_indigena:
    print("\n[Pacientes Indígenas]")
    print(df[col_indigena].value_counts())

# =========================================================
# VISUALIZACIONES
# =========================================================

sns.set_style("darkgrid")

# -------------------------
# GRÁFICA POR ESTADO
# -------------------------

if col_entidad:

    plt.figure(figsize=(12,8))

    df[col_entidad].value_counts().plot(
        kind='barh',
        color='darkcyan'
    )

    plt.title('Número de Casos por Estado', fontsize=14)
    plt.xlabel('Cantidad de Casos')
    plt.ylabel('Entidad Federativa')

    plt.tight_layout()
    plt.show()

# -------------------------
# HISTOGRAMA DE EDADES
# -------------------------

if col_edad:

    plt.figure(figsize=(10,5))

    sns.histplot(
        df[col_edad],
        bins=40,
        kde=True,
        color='firebrick'
    )

    plt.title('Distribución de Edades', fontsize=14)
    plt.xlabel('Edad')
    plt.ylabel('Frecuencia')

    plt.show()

# -------------------------
# GRÁFICA TEMPORAL
# -------------------------

if col_fecha:

    plt.figure(figsize=(12,5))

    df[col_fecha] = pd.to_datetime(df[col_fecha])

    casos_tiempo = df.groupby(col_fecha).size()

    casos_tiempo.plot(
        kind='line',
        color='navy',
        linewidth=2
    )

    plt.title('Evolución Temporal de Casos COVID-19', fontsize=14)
    plt.xlabel('Fecha')
    plt.ylabel('Casos Registrados')

    plt.show()

# -------------------------
# BARRAS APILADAS
# -------------------------

if col_sexo and col_clasificacion:

    tabla_sexo = pd.crosstab(
        df[col_sexo],
        df[col_clasificacion]
    )

    tabla_sexo.plot(
        kind='bar',
        stacked=True,
        figsize=(10,6),
        colormap='plasma'
    )

    plt.title('Distribución de Resultados por Sexo', fontsize=14)
    plt.xlabel('Sexo')
    plt.ylabel('Total de Casos')

    plt.legend(title='Clasificación')
    plt.show()

# -------------------------
# GRÁFICA CIRCULAR
# -------------------------

if col_paciente:

    plt.figure(figsize=(7,7))

    df[col_paciente].value_counts().plot(
        kind='pie',
        autopct='%1.1f%%',
        colors=['lightgreen', 'tomato'],
        startangle=140
    )

    plt.title('Pacientes Ambulatorios vs Hospitalizados', fontsize=14)

    plt.ylabel('')
    plt.show()

# -------------------------
# COMORBILIDADES
# -------------------------

comorbilidades = [
    'DIABETES',
    'HIPERTENSION',
    'OBESIDAD',
    'EPOC',
    'ASMA',
    'INMUSUPR',
    'CARDIOVASCULAR',
    'RENAL_CRONICA',
    'TABAQUISMO'
]

comorbilidades_presentes = [c for c in comorbilidades if c in df.columns]

if comorbilidades_presentes:

    plt.figure(figsize=(11,5))

    conteo_comorbilidades = [
        df[c].value_counts().get(1, 0)
        for c in comorbilidades_presentes
    ]

    sns.barplot(
        x=comorbilidades_presentes,
        y=conteo_comorbilidades,
        palette='viridis'
    )

    plt.title('Frecuencia de Comorbilidades', fontsize=14)
    plt.xlabel('Comorbilidad')
    plt.ylabel('Frecuencia')

    plt.xticks(rotation=20)

    plt.show()

# =========================================================
# PROBABILIDADES
# =========================================================

print("\n" + "="*60)
print(" PROBABILIDADES DE COMORBILIDADES ")
print("="*60)

total_registros = len(df)

for c in comorbilidades_presentes:

    casos_con_enfermedad = df[c].value_counts().get(1, 0)

    prob = casos_con_enfermedad / total_registros

    print(f"Probabilidad de padecer {c}: {prob:.6f} ({prob*100:.2f}%)")

# =========================================================
# DISTRIBUCIÓN NORMAL
# =========================================================

print("\n" + "="*60)
print(" CÁLCULO DE PROBABILIDAD (DISTRIBUCIÓN NORMAL)")
print("="*60)

mu = 2360
sigma = 714

lim_inf = 2000
lim_sup = 3000

cdf_inferior = stats.norm.cdf(lim_inf, mu, sigma)

cdf_superior = stats.norm.cdf(lim_sup, mu, sigma)

probabilidad_rango = cdf_superior - cdf_inferior

print(f"La probabilidad de que un estado tenga entre 2000 y 3000 casos es: {probabilidad_rango:.6f}")

print(f"Expresado en porcentaje: {probabilidad_rango*100:.2f}%")

# =========================================================
# INTERPRETACIÓN GENERAL
# =========================================================

print("\n" + "="*60)
print(" INTERPRETACIÓN GENERAL ")
print("="*60)

print("""

Los resultados muestran diferencias importantes
entre estados respecto al número de casos COVID.

También se observa una presencia considerable de
comorbilidades como diabetes, obesidad e hipertensión.

La mayoría de los pacientes registrados fueron
ambulatorios.

Las gráficas permiten visualizar mejor la distribución
de edades, evolución temporal y comportamiento de
los datos.

""")

# =========================================================
# CONCLUSIÓN PERSONAL
# =========================================================

print("\n" + "="*60)
print(" CONCLUSIÓN PERSONAL ")
print("="*60)

print("""

Este proyecto ayudó a comprender cómo aplicar
Python en el análisis estadístico de datos reales.

También permitió utilizar herramientas como pandas,
matplotlib y seaborn para generar estadísticas
y visualizaciones.

""")
```
