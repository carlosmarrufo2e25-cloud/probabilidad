```python
# Importamos las librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el archivo CSV (ajusta la ruta si es necesario)
covid = pd.read_csv('covid.csv')
```

```python
# Mostramos los primeros 5 registros del archivo
covid.head()
```

```python
# Mostrar la información del archivo
# Muestra número de registros, tipos de datos, valores nulos, memoria utilizada, etc.
covid.info()
```

```python
# Mostrar la edad máxima de la base de datos
covid.EDAD.max()
```

```python
# Mostrar la edad mínima de la base de datos
covid.EDAD.min()
```

```python
# Calcular la media de la edad
media = covid.EDAD.mean()

# Dibujar el histograma
plt.hist(covid.EDAD, bins=11)
plt.title("Histograma Edad Pruebas COVID")
plt.axvline(media, color='y', linestyle='dashed', linewidth=2, label=f'Media = {media:.2f}')
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.legend()
plt.show()
```

```python
# Histograma de la columna SEXO
plt.hist(covid.SEXO, bins=3)
plt.title("Histograma Sexo de Pruebas COVID")
plt.xlabel("Sexo")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Histograma del tipo de paciente
plt.hist(covid.TIPO_PACIENTE, bins=3)
plt.title("Histograma de tipos de Pacientes en Pruebas COVID")
plt.xlabel("Tipo de Paciente")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Boxplot de edad por sexo
sns.boxplot(x="SEXO", y="EDAD", data=covid)
plt.show()
```

```python
# Boxplot de edad por tipo de paciente
sns.boxplot(x="TIPO_PACIENTE", y="EDAD", data=covid)
plt.show()
```

```python
# Boxplot de edad por resultado
sns.boxplot(x="RESULTADO", y="EDAD", data=covid)
plt.show()
```
