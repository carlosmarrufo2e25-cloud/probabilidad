```python
import random
import statistics as stats
edades=[]
for i in range(100):
  na=random.randint(18,70)
  edades.append(na)
print(edades)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(edades, columns=['Edad'])
df.head()
```

```python
tabla_fecuancias= df['Edad']. value_counts().reset_index()
tabla_fecuancias.columns = ['Edad', 'Frecuancias']
tabla_fecuancias = tabla_fecuancias.sort_values(by='Edad')
print("tabla de distribucion de frecuancias")
tabla_fecuancias
```

```python
tabla_fecuancias['frec relativa'] = tabla_fecuancias['Frecuancias'] / 100
tabla_fecuancias
```

```python
tabla_fecuancias['frec porcentual'] = tabla_fecuancias['Frecuancias'] / 100 * 100
tabla_fecuancias
```
